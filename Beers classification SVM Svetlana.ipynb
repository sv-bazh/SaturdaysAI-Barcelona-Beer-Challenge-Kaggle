```python
#import xgboost as xgb
from sklearn.model_selection import train_test_split
from catboost import CatBoostRegressor
from catboost import CatBoostClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import AdaBoostRegressor
#import lightgbm as lgb
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
df_it = df.iloc[:,0:16]
imp = IterativeImputer(max_iter=10, verbose=0)
imp.fit(df_it)
imputed_df = imp.transform(df_it)
imputed_df = pd.DataFrame(imputed_df, columns=df_it.columns)
```

```python
df_it.head()
```

```python
imputed_df.head()
```

```python
df = imputed_df
```

```python
df.head()
```

```python
features = df.columns.to_list()[:15]
x = df.loc[:, features].values
x_std = StandardScaler().fit_transform(x) # normalizing the features
```

```python
round(np.mean(x_std),1),np.std(x_std)
```

```python
feat_cols = ['feature'+str(i) for i in range(x_std.shape[1])]
normalised_dataset = pd.DataFrame(x_std,columns=feat_cols)
normalised_dataset.head()
```

```python
X = normalised_dataset
y = df['Style']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
